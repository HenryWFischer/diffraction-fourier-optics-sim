```python
#imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
#parameters

#red laser (meters)
wavelength = 633e-8

#slit width (meters)
a = 0.1e-3

#distance from screen (meters)
L = 1

#screen coordinates, 20 cm wide (meters)
x = np.linspace(-0.1, 0.1, 2000)

#distance between slits (meters)
d = 0.3e-3
```

```python
#equations

beta = (np.pi*a*x)/(wavelength*L)
beta[beta==0] = 1e-20 #avoid divide by 0

#single slit
I = (np.sin(beta)/beta)**2

#double slit
I_double = I*(np.cos(np.pi*d*x/(wavelength*L)))**2
```

```python
#plots

#single slit plot
plt.figure(figsize=(10,5))
plt.plot(x*1e3, I, color = 'blue') #convert m to mm
plt.title("Single-Slit Diffraction Pattern")
plt.xlabel("Position to Screen (mm)")
plt.ylabel("Normalized Intensity")
plt.grid(True)
plt.show()

#double slit plot
plt.figure(figsize=(10,5))
plt.plot(x*1e3, I_double, color = 'red') #convert m to mm
plt.title("Double-Slit Diffraction Pattern")
plt.xlabel("Position to Screen (mm)")
plt.ylabel("Normalized Intensity")
plt.grid(True)

plt.show()
```
